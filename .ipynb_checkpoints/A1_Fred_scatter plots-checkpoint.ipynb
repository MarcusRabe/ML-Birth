```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn import linear_model
import time
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsRegressor # KNN for Regression
import statsmodels.formula.api as smf # regression modeling
import xgboost as xgb
import graphviz
from sklearn.linear_model import LassoCV, LassoLarsCV, LassoLarsIC, MultiTaskLasso, Lasso, ElasticNet, lasso_path, enet_path
%matplotlib inline
```

```python
file = "birthweight_feature_set.xlsx"
birthweight = pd.read_excel(file)
```

```python
for col in birthweight:

    """ Create columns that are 0s if a value was not missing and 1 if
    a value is missing. """
    
    if birthweight[col].isnull().any():
        birthweight['m_'+col] = birthweight[col].isnull().astype(int)

birthweight_dropped = birthweight.dropna()
```

```python
birthweight['meduc'] = birthweight['meduc'].fillna(birthweight_dropped['meduc'].mean())
birthweight['meduc'] = birthweight['meduc'].round(0)
```

```python
birthweight['npvis'] = birthweight['npvis'].fillna(birthweight_dropped['npvis'].mean())
birthweight['npvis'] = birthweight['npvis'].round(0)
```

```python
birthweight['feduc'] = birthweight['feduc'].fillna(birthweight_dropped['feduc'].mean())
birthweight['feduc'] = birthweight['feduc'].round(0)
```

```python
birthweight['sameEduc'] = (birthweight['meduc'] == birthweight['feduc'])*1
birthweight['bothBlck'] = (birthweight['mblck'] * birthweight['fblck'])
birthweight['bothWhte'] = (birthweight['mwhte'] * birthweight['fwhte'])
birthweight['bothOth'] = (birthweight['moth'] * birthweight['foth'])
birthweight['raceSame'] = birthweight['bothBlck'] + birthweight['bothWhte'] + birthweight['bothOth']
```

```python
birthweight.columns.values.tolist()
```

```python
# attributes
X = birthweight.drop(['bwght', 'omaps', 'fmaps'], 1)
# lables
y = birthweight['bwght']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=508)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
score = lr.score(X_test, y_test)
train_score = lr.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
X_n = X.drop(['meduc', 'feduc', 'mwhte', 'mblck', 'moth', 'fwhte', 'fblck', 'foth', 'sameEduc'],1)

X_train_n, X_test_n, y_train, y_test = train_test_split(X_n, y, test_size=0.1, random_state=508)
```

```python
lr = LinearRegression()
lr.fit(X_train_n, y_train)
```

```python
score = lr.score(X_test_n, y_test)
train_score = lr.score(X_train_n, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
reg = linear_model.Ridge(alpha = 150,
                         random_state = 508,
                         normalize = False)
reg.fit(X_train, y_train)
```

```python
score = reg.score(X_test, y_test)
train_score = reg.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
regL = linear_model.Lasso(alpha=30)
regL.fit(X_train, y_train)
```

```python
score = regL.score(X_test, y_test)
train_score = regL.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
regLCV = linear_model.LassoCV(cv=50, random_state=508)
regLCV.fit(X_train, y_train)
```

```python
score = regLCV.score(X_test, y_test)
train_score = regLCV.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
regEN = ElasticNet(alpha=2.5, l1_ratio=0.7)
regEN.fit(X_train, y_train)
```

```python
score = regEN.score(X_test, y_test)
train_score = regEN.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
regLARS = linear_model.LassoLars(alpha=2, max_iter = 2000)
regLARS.fit(X_train, y_train)
```

```python
score = regLARS.score(X_test, y_test)
train_score = regLARS.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 51)
for n_neighbors in neighbors_settings:
    # Building the model
    clf = KNeighborsRegressor(n_neighbors = n_neighbors)
    clf.fit(X_train, y_train)
    
    # Recording the training set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    
    # Recording the generalization accuracy
    test_accuracy.append(clf.score(X_test, y_test))

# Plotting the visualization
fig, ax = plt.subplots(figsize=(12,9))
plt.plot(neighbors_settings, training_accuracy, label = "training accuracy")
plt.plot(neighbors_settings, test_accuracy, label = "test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend()
plt.show()
```

```python
print("The optimal number of neighbors is", \
      test_accuracy.index(max(test_accuracy)), \
      "with an optimal score of", \
      max(test_accuracy))
```

```python
knn_reg = KNeighborsRegressor(algorithm = 'auto',
                              n_neighbors = 3)
knn_reg.fit(X_train, y_train)
```

```python
score = knn_reg.score(X_test, y_test)
train_score = knn_reg.score(X_train, y_train)
diff = train_score - score

print("Testing score: %f" % (score))
print("Training score: %f" % (train_score))
print("Diff: %f" % (diff))
```

```python
xg_reg = xgb.XGBRegressor(objective ='reg:linear',
                          colsample_bytree = 1,
                          learning_rate = 0.3,
                          max_depth = 20,
                          alpha = 0.1,
                          n_estimators = 11,
                          subsample = 1)
xg_reg.fit(X_train,y_train)
```
